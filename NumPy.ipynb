Упр. - Создание массивов NumPy (ndarray)

```python
# Из списка Python
arr1 = np.array([1, 2, 3, 4, 5])



# Массив из нулей
arr2 = np.zeros(5)



# Массив из единиц
arr3 = np.ones((2, 3))



# Диапазон значений
arr4 = np.arange(0, 10, 2)
```

Упр. - Свойства массивов NumPy (ndarray)

```python
arr = np.array([[1, 2, 3], [4, 5, 6]])

print("Размерность:", arr.ndim)       # Вывод: 2

print("Форма:", arr.shape)          # Вывод: (2, 3)

print("Тип данных:", arr.dtype)     # Вывод: int64 (зависит от платформы)
```

Упр. - Арифметические операции (ndarray)

```python
arr1 = np.array([1, 2, 3])

arr2 = np.array([4, 5, 6])



print(arr1 + arr2)  # Вывод: [5 7 9]

print(arr1 * 2)     # Вывод: [2 4 6]
```

Упр. - Индексация и срезы (ndarray)

```python
arr = np.array([0, 1, 2, 3, 4, 5])

print(arr[0])        # Вывод: 0

print(arr[1:4])     # Вывод: [1 2 3]

print(arr[::-1])    # Вывод: [5 4 3 2 1 0] (обратный порядок)
```

Упр. - Векторизация: ускорение вычислений с помощью NumPy

```python
# Без векторизации (медленно)

arr = np.arange(1, 1000000)

result = []

for i in arr:

    result.append(i**2)

# С векторизацией (быстро)

result = arr**2
```

```python
height = np.array([1.68, 1.78, 1.65])  # Рост в метрах

weight = np.array([54, 71, 60])  # Вес в килограммах

bmi = weight / (height ** 2)

print(bmi)
```
