```python
# Из CSV файла
df = pd.read_csv('data.csv')



# Из Excel файла
df = pd.read_excel('data.xlsx', sheet_name='Sheet1')



# Из JSON файла
df = pd.read_json('data.json')
```

Группировка по столбцу 

```python
data = pd.read_csv('C:/Users/123/Downloads/sales.csv')
```

```python
data
```

```python
grouped = data.groupby('ID товара')['Цена'].mean()
grouped.head()
```

Применение нескольких функций агрегации

```python
grouped = data.groupby('ID товара').agg({'Цена': ['mean', 'sum'], 'Скидка': 'max'})
grouped.head()
```
